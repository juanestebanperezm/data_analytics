```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np 


df=pd.read_csv('top10s.csv',index_col=0,encoding = "ISO-8859-1")

df.head(10)
```

<h1>Cual año tuvo mas lanzamientos 🤔</h1>


```python

#Año y titulos
year_title=df[['year','title']].groupby('year').count().reset_index()

sns.relplot(x='year',y='title',data=year_title,kind='line',aspect=2)

plt.title('Canciones x Año')
plt.xlabel('Año')
plt.ylabel('Titulo')
plt.show()
```

<h1>Distribucion de canciones x genero  🤠</h1>


```python
plt.clf()

datos=df[['top genre','title']].groupby('top genre').count().reset_index()

sns.catplot(x='top genre',y='title',data=datos,kind='bar',aspect=3,color='blue')

plt.title('Distribucion de canciones x genero')
#Espaciados entre los datos del Eje X para asi evitar que colapsen los datos descriptivos :) 
plt.xticks(rotation=90)
plt.xlabel('Top genero')
plt.ylabel('Numero de canciones')
plt.show()
```

<h1>Popularidad de Eminem </h1>

```python
plt.clf()
datos_=df[df['artist']=='Eminem']


sns.relplot(x='title',y='pop',data=datos_,kind='line',aspect=3,palette='Set2')

plt.xticks(rotation=90)
plt.title('Popularidad de Eminem')
plt.xlabel('Cancion')
plt.ylabel('Popularidad')
plt.show()
```

<h1>Popularidad de Rihanna<h1>

```python
plt.clf()
datos_riri=df[df['artist']=='Rihanna']

sns.relplot( x='title',y='pop',data=datos_riri,aspect=3,kind='line',palette='Dark2' )
plt.xticks(rotation=90)
plt.xlabel('Titulo')
plt.ylabel('Popularidad')
plt.show()
```

<h1>Popularidad de los generos :) <h1>

```python
plt.clf()

datos_generos=df[['top genre','pop']].groupby('top genre').mean().reset_index()

sns.catplot(
            x='top genre',
            y='pop',
            data=datos_generos,kind='point',
            aspect=2 ,
            color='purple')


plt.title('Popularidad de generos')
plt.xticks(rotation=90)
plt.xlabel('Top genero')
plt.ylabel('Popularidad')
plt.show()
```

<h1>Valencia de los artistas</h1>

```python
plt.clf()
sns.relplot(x='year',y='val',data=df,kind='line',aspect=3)
plt.title('Año de lanzamiento')
plt.xlabel('Año')
plt.ylabel('Valencia')
plt.show()
```

<h1>Popularidad de genero de artistas</h1>

```python
plt.clf()

datos_pop=df[df['top genre']=='pop']

sns.catplot(
            x='artist',
            y='pop',
            data=datos_pop,
            kind='box',
            aspect=2,
            palette='Dark2'
            )
plt.title('Generos mas populares de los artistas')
plt.xticks(rotation=90)
plt.xlabel('Artista')
plt.ylabel('Popularidad')
plt.show()
```

```python
nuevo=df[['artist','pop']].groupby('artist').mean().astype(int).reset_index().sort_values(by='pop',ascending=0).head(10)

sns.catplot(
    x='artist',
    y='pop',
    data=nuevo,kind='bar',
    aspect=3,
    palette='Dark2'
    )
plt.title('Popularidad de artista')
plt.xticks(rotation=90)
plt.xlabel('artista')
plt.ylabel('popularidad')
plt.show()
```
