```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('../input/consumer-feedback-and-electronic-products-a-datae/DatafinitiElectronicsProductData.csv',index_col=0)

df.dropna()
```

**categories by Brand**


So here's a visual representation of most categories of electronic products that has a brand on their portfolio haha, sorry my english isn't too good, but you can understand the plot

```python
cat_by_mark=df[['brand','categories']].groupby('brand').count().reset_index().sort_values(by='categories',ascending=False).head(10)

sns.catplot(
    data=cat_by_mark,
    x='brand',
    y='categories',
    aspect=3,
    palette='Dark2',
    kind='bar')

plt.title('Categories By Brand')
plt.xticks(rotation=90)
plt.xlabel('Brand')
plt.ylabel('Num of Categories')
plt.show()
```

**Correlation Heatmap**

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True,cmap="YlGnBu")
```

**What are the most common colors at the categories**

```python
cat_color=df[['categories','colors']].groupby('colors').count().sort_values(by='categories',ascending=False).reset_index().head(10)
sns.catplot(
    x='colors',
    y='categories',
    kind='bar',
    aspect=3,
    data=cat_color,
    palette='Set2')
plt.title('Most common colors at electronics')
plt.xticks(rotation=90)
plt.ylabel('Categories')
plt.xlabel('Colors')
plt.show()
```

**Top 10 Recomended Brands**

```python

recomended_brands=df[['brand','reviews.doRecommend']].dropna().groupby('brand').mean().sort_values(by='reviews.doRecommend',ascending=False).reset_index().head(10)
plt.figure(figsize=(20,20))
sns.catplot(data=recomended_brands,x='brand',y='reviews.doRecommend',kind='bar',aspect=3)
plt.xticks(rotation=90)
plt.title('Top 10 recomended brands')
plt.xlabel('brand')
plt.ylabel('REVIEWS count')
plt.show()
```

**Categories with less rating**

```python
rating_product=df[['categories','reviews.rating']].groupby('categories').mean().sort_values(by='reviews.rating',ascending=1).head(5)

sns.relplot(data=rating_product,x='categories',y='reviews.rating',kind='line',aspect=3)
plt.xticks(rotation=90)
plt.title('categories with less rating')
plt.ylabel('rating')
plt.xlabel('category')
plt.show()
```
