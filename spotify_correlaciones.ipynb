```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn


df=pd.read_csv('top10s.csv',encoding = "ISO-8859-1",index_col=0)
df.describe()
```

<h1>Ploteo de todos los datos:Viste general de estos </h1>

```python
fig=plt.figure(figsize=(15,10))
ax=fig.gca()
df.hist(ax=ax)
plt.style.use('ggplot')
```

<h1>Ploteo de correlacion entre diferentes caracteristicas del dataset <h1>

```python
grafic=sns.pairplot(df,y_vars='pop',
                    x_vars=['bpm', 'nrgy', 'dnce', 'dB','live', 'val', 'dur', 'acous', 'spch'],
                    palette='Set2')

grafic.fig.set_figheight(3)
grafic.fig.set_figwidth(20)
```

<h1>Mapa de calor de las correlaciones<h1>

```python
plt.figure(figsize=(20,20))

sns.heatmap(df.corr(),annot=True,cmap="YlGnBu")
```
